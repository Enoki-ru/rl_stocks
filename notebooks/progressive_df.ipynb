```python
import random
import sys
import os
from pathlib import Path

# Это нужно сделать ОДИН РАЗ в начале ноутбука
# Находим корень проекта (на два уровня выше текущей директории ноутбука)
# и добавляем папку src в пути для импорта
project_root = Path.cwd().parent 
src_path = project_root / 'src'
if str(src_path) not in sys.path:
    sys.path.append(str(src_path))

from local_paths import CONFIGS_DIR
import pandas as pd
import yaml


CONFIGS_DIR
config_name = 'light.yaml'

with open(CONFIGS_DIR / config_name, 'r') as file:
    config_params = yaml.safe_load(file)

def generate_params(config, param_name_cnt_steps: str="max_steps"):
    n_steps = config.get(param_name_cnt_steps, 1)
    keys_cfg = config['progress']
    step = 0

    while True:
        t = min(step / (n_steps - 1), 1.0) if n_steps > 1 else 1.0
        current_payload = {}
        for key, limits in keys_cfg.items():
            start, end = limits
            low = int(start[0] + (end[0] - start[0]) * t)
            high = int(start[1] + (end[1] - start[1]) * t)
            current_payload[key] = random.randint(low, high)
            
        yield current_payload
        step += 1

q = generate_params(config_params, param_name_cnt_steps='test_max_steps')

print(f"--- Генерация параметров в динамических интервалах ---")
for i in range(50):
    print(f"#{i+1}:", next(q))
```
